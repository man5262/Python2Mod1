## Exercise 1: 
Create a Player class with a name, a position, and a speed attributes. Set the initial value of speed to 1 and create any necessary methods.
Then, 
1. Create two Player objects and print information about both players.
2. Assign 8 to player 1's speed
3. print information about both players.

```python
class Player:
    """This is a Player class"""
    def __init__(self, name, position):
        """Initialization method"""
        self.name = name
        self.position = position
        self.speed = 1
    
    def set_speed(self, speed):
        self.speed = speed
        
    def print_player_info(self):
        return f"Name:\t\t{self.name}\nPosition:\t{self.position}\nSpeed:\t\t{self.speed}"
        
player1 = Player("Sam", "First")
player2 = Player("Bob", "Second")

# Printing newly created objects' information
print(f"Initial Player 1 info:\n\n{player1.print_player_info()}\n")
print(f"Initial Player 2 info:\n{player2.print_player_info()}\n\n\n")

# Changing Player's 1 speed property and print both objects info again
player1.set_speed(8)

# Printing objects' information after update
print(f"Initial Player 1 info:\n\n{player1.print_player_info()}\n")
print(f"Initial Player 2 info:\n{player2.print_player_info()}\n\n\n")
```

## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

```python
class Rectangle():
    """This is a Rectangle class"""
    def __init__(self, height, length):
        """Rectangle Class Initialization"""
        self.height = height
        self.length = length
        
    def perimeter(self):
        return (self.height + self.length) * 2
        
    def area(self):
        return self.height * self.length
    
    def print_rectangle_info(self):
        return f"Length:\t{self.length}\nHeight:\t{self.height}\n"
    
rect1 = Rectangle(9, 10)
rect2 = Rectangle(8, 11)
print(f"Initial Rectangle 1 info:\n\n{rect1.print_rectangle_info()}")
print(f"Initial Rectangle 2 info:\n\n{rect2.print_rectangle_info()}")

# Analysing rectangle's perimeters
p1 = rect1.perimeter()
print(f"Perimeter of rectangle 1 is {p1}")
p2 = rect2.perimeter()
print(f"Perimeter of rectangle 2 is {p2}")
if (p1 > p2):
    print("Rectangle 1 has larger perimeter")
elif (p1 < p2):
    print("Rectangle 2 has larger perimeter")
else:
    print("Both perimeters are equal\n")

# Analysing rectangle's areas
a1 = rect1.area()
print(f"Area of rectangle 1 is {a1}")
a2 = rect2.area()
print(f"Area of rectangle 2 is {a2}")
if (a1 > a2):
    print("Rectangle 1 has larger area")
elif (a1 < a2):
    print("Rectangle 2 has larger area")
else:
    print("Both areas are equal\n")

```

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



```python
import math

class Triangle():
    """Initialization method for right triangle class"""
    def __init__(self, height, base):
        self.height = height
        self.base = base
        
    def perimeter(self):
        return self.height + self.base + math.sqrt(self.height ** 2 + self.base ** 2)
        
    def area(self):
        return self.height * self.base / 2
    
    def print_triangle_info(self):
        return f"Height:\t{self.height}\nBase:\t{self.base}\n"
    
tri1 = Triangle(3, 4)
tri2 = Triangle(6, 8)
print(f"Initial Triangle 1 info:\n\n{tri1.print_triangle_info()}")
print(f"Initial Triangle 2 info:\n\n{tri2.print_triangle_info()}")

# Analysing triangle's perimeters
p1 = tri1.perimeter()
print(f"Perimeter of triangle 1 is {p1}")
p2 = tri2.perimeter()
print(f"Perimeter of triangle 2 is {p2}")
if (p1 > p2):
    print("Triangle 1 has larger perimeter")
elif (p1 < p2):
    print("Triangle 2 has larger perimeter")
else:
    print("Both perimeters are equal\n")

# Analysing triangle's areas
a1 = tri1.area()
print(f"Area of triangle 1 is {a1}")
a2 = tri2.area()
print(f"Area of triangle 2 is {a2}")
if (a1 > a2):
    print("Triangle 1 has larger area")
elif (a1 < a2):
    print("Triangle 2 has larger area")
else:
    print("Both areas are equal\n")
    
    
```

```python
class Employee():
    """Initialization of Employee class"""
    def __init__(self):
        self.firstname = ""
        self.lastname = ""
        self.joined_date = 0
        self.active = False
    
    def set_active(self):
        if self.joined_date < 2000:
            self.active = False
        else:
            self.active = True

employees = []
ind_employee = Employee()
dataset = [{"first":"elmo", "last":"monster", "year": 2000},
           {"first":"count", "last":"count", "year": 1819},
           {"first":"big", "last":"bird", "year": 2011},
           {"first":"cookie", "last":"monster", "year": 2002},
           {"first":"bert", "last": "person", "year": 2013},
           {"first":"ernie", "last": "person", "year": 2013},
           {"first":"oscar", "last": "grouch", "year": 1999},
           {"first":"zoe", "last": "cutie", "year": 2015},
           {"first":"abby", "last": "cadabby", "year": 2016},
           {"first":"prairie", "last": "dawn", "year": 2020}]

# Collect all data
for dict in dataset:
    ind_employee.firstname = dict["first"]
    ind_employee.lastname = dict["last"]
    ind_employee.joined_date = dict["year"]
    ind_employee.set_active()

    employees.append(ind_employee)
    ind_employee = Employee()
    
# Print all data
if len(employees) > 0:
    print("First Name\tLast Name\tJoined\tActive?")
    for ind_class in employees:
        print (f"{ind_class.firstname}\t\t{ind_class.lastname}\t\t{ind_class.joined_date}\t{ind_class.active}")
        
# Print only active employees
if len(employees) > 0:
    print("\nFirst Name\tLast Name\tJoined\tActive?")
    for ind_class in employees:
        if ind_class.active:
            print (f"{ind_class.firstname}\t\t{ind_class.lastname}\t\t{ind_class.joined_date}\t{ind_class.active}")
            
```
